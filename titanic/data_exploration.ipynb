```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
def display_barchart_gender(data: pd.DataFrame):
    
    male = df['Sex'] == 'male'
    female = df['Sex'] == 'female'
    survived = df['Survived'] == 1
    died = df['Survived'] == 0
    men = (
        len(df[male]),
        len(df[male & died]),
        len(df[male & survived])
    )
    women = (
        len(df[female]),
        len(df[female & died]),
        len(df[female & survived])
    )

    index = np.arange(len(men))
    width = 0.40

    fig, ax = plt.subplots(figsize=(10,10))
    p1 = ax.bar(index, men, width, label='Men')
    p2 = ax.bar(index, women, width, bottom=men, label='Women')

    ax.set_ylabel('Number of people')
    ax.set_title('Survival and Death Based on Gender')
    ax.set_xticks(index)
    ax.set_xticklabels(['Total on board', 'Died', 'Survived'])
    ax.legend()

    ax.bar_label(p1, label_type='center')
    ax.bar_label(p2, label_type='center')
    ax.bar_label(p2)

    plt.show()

    
display_barchart_gender(df)
```
